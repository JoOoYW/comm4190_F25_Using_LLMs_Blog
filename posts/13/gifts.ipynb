## My Dilemma

I have two very close friends, one currently resides in Virginia and the other in Pittsburgh. They are both six hours away in opposite directions. How am I supposed to shop for them? And UPenn's so stingy with their breaks that they start their breaks earlier than me, and I work the few days that I even am home! What can I do? But never fear, my 21 questions and experience from our years together will help me find a perfect gift for them with the help of Perplexity.
