## The Competition 

On November 23 at 12 PM, one of the TAs for CIS 1600 posted on my [edstem.org](edstem.org/us/dashboard) about the annual CIS 1600 merch design contest. Now, I like desiging things, but to design one for a STEM course, much less one of THE nerdiest courses you can take is a challenge. It's hard to make it look good and "hip". So what's to do? 

## A General Theme 

Unfortunately, it's not easy to make a STEM related t-shirt without making it cringey. Most, if not all of these shirts, are jokes on a graphic tee in a duotone color scheme.
