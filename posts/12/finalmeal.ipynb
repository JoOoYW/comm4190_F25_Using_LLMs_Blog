## Context 
The origin of a final meal has not been well documented, but since the 1891 there was great interest in the idea. Believed to have religious and historical roots, with the Last Supper in Christianity being a significant early example, the practice of granting customized final meals to those on death row is a more recent, and peculiarly American, fascination. 

But over the years, there have been restrictions and some states, like Texas, have stopped the practice of customized last meals. The reason for this is understandable as in 2011, Lawrence Russell Brewer requested a [grandiose meal](https://www.theguardian.com/world/2011/sep/23/texas-execution-ends-final-meal) and then refused to eat it. The uneaten meal caused public outcry and a strong reaction from state officials, particularly Senator John Whitmire, who called the privilege "extremely inappropriate". 

But enough about that, what would I have as my last meal?
