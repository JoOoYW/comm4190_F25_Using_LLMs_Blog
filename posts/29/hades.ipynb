## A basic strategy

### Core priorities

- Learn enemy patterns and stop attacking briefly when you see a new enemy or boss move; dodging safely matters more than squeezing in every hit.
- Constantly move and dash, treating your dash as your main defensive tool; early on, prioritize mobility and survivability over raw damage.
- Expect to die a lot and treat each run as resource farming for mirror upgrades, weapon aspects, and keepsakes rather than “failure”.
### Mirror and resource strategy

- Spend darkness on extra dash, Death Defiance, and Chthonic Vitality / healing first, since these give you the biggest margin for error on real clears.
- Use the weapon that glows with a darkness bonus to farm resources faster, even if it’s not your favorite, until your core mirror talents are upgraded
- Use nectar once on key NPCs (especially Olympian gods and Skelly) to unlock strong keepsakes that boost survivability or give you guaranteed boons.

### Boons and builds that carry

- Prioritize defensive/mobility boons like Athena’s Divine Dash, Hermes extra dash and speed, and strong calls like Athena’s or Aphrodite’s Aid; they are widely recommended for surviving the final boss.
- Build around a few powerful synergies instead of grabbing random damage: Athena + Ares (Merciful End), Zeus lightning builds, or Artemis crit builds are all known to trivialize many encounters when they come together.
- For comfort, many players favor ranged-leaning setups (bow, cast-focused, or safe shield builds) so they can keep distance from dangerous enemies and Hades himself.

### Final boss habits

- Treat Hades like a pattern-learning fight: keep your distance, circle him, and only step in when his current attack pattern leaves a clear opening.
- Use pillars in the arena to block his projectiles and especially his beam/flamethrower in phase two; hiding behind terrain makes those attacks almost trivial.
- Do not panic-dash around randomly; many experienced players recommend dashing into or through him to avoid wide swings, and using shield block or Athena dash to neutralize big attacks.

### A simple game plan per run

- Pick the darkness-bonus weapon unless you are on a serious “clear attempt,” in which case choose the weapon you play most confidently.
- Use a survivability keepsake early (extra health, damage reduction, or Athena/Aphrodite keepsakes for guaranteed defensive boons), then swap later to a damage keepsake for Elysium and Styx.
- In rooms, play patient: clear small threats first, stay mobile, and only greed damage when the room is under control; treat boss rooms and the final fight as pattern drills rather than DPS races.

## An extensive inspection

![The gods](gods.png)

### Always high‑value gods

- Athena: Divine Dash is widely considered the best boon in the game; it lets your dash deflect projectiles and massively cuts incoming damage. Her attack/special (Divine Strike/Flourish) also deflect and scale damage, and she can restore Death Defiance with Athena’s Aid tree.
- Hermes: His boons give extra dashes, move speed, and dodge chance, which make every build safer and more forgiving. Hyper Sprint / Rush Delivery and attack-speed boosts are strong on nearly any weapon.

### Gods for survivability

- Aphrodite: Weak on hit reduces enemy damage, and she offers flat damage reduction and max HP boosts, making her a top survival pick that still hits very hard. Aphrodite’s Aid is one of the strongest panic-button Calls for boss fights.
- Dionysus: Hangover (damage over time) is easy to apply while kiting, and he has boons that heal after encounters or improve fountains, helping long-term survival.

### Gods for big damage

- Artemis: Crit chance on attack/special/cast can shred bosses; boons like Exit Wounds or extra crit arrows turn certain builds into boss melters. She also pairs in strong duos like Merciful End with Ares or cast synergies with Poseidon.
- Zeus: Great for room clear with chain lightning and lightning-on-dash; becomes very strong if you stack lightning upgrades and jolted on enemies.

### “Auto-include” boons if offered

- Defensive/mobility: Divine Dash (Athena), Athena’s attack/special, extra dashes (Hermes), move speed/dodge (Hermes), damage reduction or max HP from Aphrodite/Dionysus.
- Boss tools: Aphrodite’s Aid, Athena’s Aid, or a high-level Call from Zeus/Dionysus all make Hades much safer and faster.
