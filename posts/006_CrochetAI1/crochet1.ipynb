## Crocheting 

Crochetings is defined as a "process of creating textiles by using a crochet hook to interlock loops of yarn, thread, or strands of other materials" and is derived from the French word 'crochet' meaning hook. There are many pieces techniques and steps to follow before a project, or design, can come to fruition. But the question still stands, can ChatGPT, or any chatbot really, give an answer that makes something decent? Chatbots are known to give information you have to double check yourselves in some way, shape, or form so what better way to test it than make something.
