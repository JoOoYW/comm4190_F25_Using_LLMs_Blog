## Gemini's poncho design
### Overview
This poncho is made from one big rectangle, folded and seamed to leave a neck opening.
It uses only chains and double crochet, so it is fast and easy to customize.

#### Materials
- [ ] Worsted weight yarn (#4) or similar cozy yarn, about 1000–1400 m for an adult size (exact amount depends on your size and length)
- [ ] Hook
- [ ] Scissors
- [ ] Yarn needle

#### Stitch Pattern (Body)
Work a simple double‑crochet rectangle:

Chain a length equal to your desired poncho length; this starting chain will be the short side of the rectangle.

1. Row 1: Double crochet (dc) in the 3rd chain from hook and in each chain across; turn.
2. Row 2: Chain 2 (counts as dc), dc in each stitch across; turn.
3. Repeat Row 2 until the long side of your rectangle matches your wrist‑to‑wrist measurement.

Optional texture: Alternate rows of half‑double crochet and back‑loop‑only rows for a ribbed look if you like.

#### Construction
1. Lay the rectangle flat with the long side horizontal.
2. Fold it in half so you get a smaller rectangle, lining up short sides together.
3. Along the top edge, seam from one side toward the center using slip stitch or whip stitch, but stop early to leave a neck hole 9–11 in / 23–28 cm wide, depending on preference.
4. Try it on: if the neck is too wide, seam a bit more; if too tight, undo a bit of the seam.

##### Optional Neckline & Edging
- Neck edging: Join yarn at the neck edge, work 1–2 rounds of single crochet evenly around to firm the opening; keep the stitch count loose so it doesn’t pull.
- Bottom edging: Add 1–3 rounds of single crochet or a simple shell (5 dc in one stitch, skip 2, sc, skip 2, repeat) for a pretty border.

##### Simple Style Variations
- Cozy turtleneck: Work several more rounds of single crochet or half‑double crochet around the neck to build a tall, soft collar.
- Stripes: Change color every few rows across the rectangle, carrying yarn up the side and hiding it in the seam.
- Hood: After finishing the neck, you can crochet a hood rectangle and seam the back, as in some hooded poncho designs.

## Is it a good design? 

Looking through the design, it's very minimal in it's direction. What does it mean by poncho length? Is it talking about how it drapes on the width of your shoulder or the depth of your torso? I also think that when laying it out, the corners should line up and be folded as a triangle, not as a rectangle. A poncho looks like a poncho because when you remove the person, it's one big triangle. However, I do like the variations and design choices given so you can customie it how you see fit.
