## Final Thoughts 

![](bookai.png)

I have just finished Ananthaswamy's book on the math behind AI and its history of how it got to this point. I've learned that under all the buzzwords, AI is mostly a stubborn number‑cruncher that keeps guessing, checking how wrong it is, and tweaking itself until it stops making wrong guesses on new data. Once I saw how much of it boils down to “turn the world into numbers, draw (very high‑dimensional) lines, and roll downhill on an error curve,” the whole field genuinely felt a lot less like wizardry and a lot more like patient, albeit slightly obsessive bookkeeping.

What really amused me is how long and messy the road to today’s systems has been: early perceptrons marketed as an invention 1000 years ahead of its time, then declared nearly useless, then rescued by multilayer networks, kernels, backprop, and a lot of computing power. All these grand arguments over decades, and yet the core ideas keep circling back to the same math that’s been sitting in textbooks for ages. I do wonder though, **if what is esentially a pile of linear algebra, calculus, and probability can accidentally evolve into code‑writing, image‑drawing models, what else might these “just math” systems surprise us with next?**
