## Inhale & Exhale

Alright, let's get you centered. Find a comfortable spot where you won't be disturbed for the next five minutes. You can sit in a chair with your feet flat on the floor and your back straight but not stiff, or you can sit on a cushion on the floor. Gently close your eyes.

Here we go.

### **Minute 1: Settle In and Find Your Breath**

Start by just noticing your body. Feel the weight of it on the chair or the floor. Let your hands rest comfortably in your lap. Now, bring your attention to your breath. Don't try to change it yet. Just notice it. Feel the air coming in through your nose, filling your lungs, and then the sensation of it leaving your body. In... and out. Just one breath at a time.

### **Minute 2: Anchor to the Present**

As you continue to breathe, start to count your breaths silently to yourself. Inhale, and in your mind, say "one." Exhale, "two." Inhale, "three." Exhale, "four." Keep going up to ten, and then start back at one. If your mind wanders—and it will, that's completely normal—just notice where it went, and gently guide it back to counting your breaths. No judgment. Just return to "one."

### **Minute 3: Body Scan for Release**

Now, let go of the counting. Bring your awareness down to your feet. Notice any sensations there—warmth, tingling, pressure. Slowly, move your attention up your legs, to your thighs, your hips, and your stomach. Notice any tension you're holding. As you breathe out, imagine that tension melting away, softening and releasing. Continue this scan up through your chest, your back, your shoulders (a common spot for stress), your neck, and finally your jaw and face. Let your jaw go slack and your forehead be smooth.

### **Minute 4: Visualize Calm**

With your body feeling a bit more relaxed, bring an image to mind. Picture a place where you feel completely calm and safe. It could be a quiet beach, a forest, or even just a cozy room. Imagine you are there. What do you see? What do you hear? Maybe the sound of gentle waves or the rustle of leaves. Feel the calm of this place washing over you, replacing any lingering stress with a sense of peace.

### **Minute 5: Return and Set an Intention**

Slowly, let the image fade and bring your focus back to the simple feeling of your breath moving in and out. Take one deep, deliberate breath in, filling your belly with air. And a long, slow breath out. When you're ready, start to wiggle your fingers and toes. Gently open your eyes, taking a moment to notice the space around you. Before you jump back to studying, set a simple intention: to carry this feeling of calm with you, and to remember you can always return to your breath for a moment of peace.

You've got this. Five minutes can make a real difference. Good luck with your finals.
