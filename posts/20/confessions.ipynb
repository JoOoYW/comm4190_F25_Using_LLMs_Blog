## My confession 

I am currently a first year studying computer science at one of the eight Ivy League schools. It's not a surprise that there are insanely smart and talented students everywhere. But I can't help feel that I shouldn't be here, like there is a very big feeling of inadequacy in my chest. I confided in one of the upper classmen who  said it was normal, and all of my professors tell us that we are here for a reason. I just feel like my reason for being here is not that strong in comparison to others (which I shouldn't be doing but I can't help it). My mind tells me I'm experiencing imposter syndrome, one of the two common feelings first years can have at college (seasonal depression is the other one).
