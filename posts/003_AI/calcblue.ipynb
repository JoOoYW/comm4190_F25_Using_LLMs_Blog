## How does Professor G's CalcBlue GPT work?

### Professor G, or Professor Robert Ghrist, has given students of MATH 1410 a GPT to help with their understanding 

Based on his book with the same title, Prof G fed ChatGPT every problem and lesson plan covered for the semester. For those taking MATH 1400 or MATH 1410, Calc Star Green/Blue has proven incredibly useful and helpful. 

The commands available include /intro, /ex, /score, and /test. Calling them for their purpose is fairly intuitive, while some like /score is when the user wants their test to be rated by an emoji or /go for a week's specific material. 

However, when prompted to give information about Week Four, as of September 14 it limits itself to four of the nine learning objectives. It decided Week 4 was "all about determinants as volume-scaling factors". When prompted about the rest of the learning objectives, it corrected itself and revised the response it gave to my original question for help.
