#### Ingredients 
* [ ] 500g AP flour
* [ ] 300mL lukewarm water
* [ ] 10g salt
* [ ] 3 bunches of green scallions (white and green)
* [ ] 40mL of high smoke point oil (we use peanut oil)

#### Steps
##### Dough:
1. Mix the flour, salt, and water together in a large mixing bowl
2. Once combined sufficiently, move it to a surface with flour and knead until smooth
3. form a round ball
4. Grab your choice of oil and put some onto the surface
5. Let the dough bowl sit for an hour

##### Layers
To form the layers inside of the scallion pancake, you need to add a mixture of oil and flour
1. Mix the oil and some flour into a small mixing bowl
2. Take a small ball of dough and flatten it out
3. Make it into a circle
4. Add the oil mixture down the middle, avoiding the sides
5. Sprinkle the chopped green scallions onto the oil
6. Fold the sides into the middle so it's one long strip
7. Roll it up into a cinnamon roll and pinch the end into the dough
8. Roll it out again into a circle
9. Repeat step 8 two more times

##### Frying 
1. Add enough oil into the pan and heat it up
2. Drop a pancake and fry until both sides are golden brown, or at least cooked
3. Enjoy!
